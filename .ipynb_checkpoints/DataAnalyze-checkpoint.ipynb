```python
def getAngle(cnt, cycle = 11) :
    
    n = (cnt%2 * cycle//2 + cnt//2 + cnt%2) % cycle
    angle = (360/cycle)*n
    
    return angle
```

```python
FileName = "2021_07_09_11_58_30_Practice_800_50_logs" # Edit Here to Change File Root
```

```python
# Load DataFrame

df = pd.read_csv(fr"C:/Users/USER-PC/Desktop/DualSensorMouse_Fittslike_Line/Logs/{FileName}/Pos_values.csv")
df["Angle"] = df["Count"].map(lambda x : getAngle(x))
```
